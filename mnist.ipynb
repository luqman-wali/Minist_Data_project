```python
from keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train=x_train/255
```

```python
x_train[4]
```

**Create CNN model:**
two conventional layer:
two polling layer:

```python
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
```

**Create Model Bulding**

```python
 model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary ()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
x_test = x_test / 255
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)
```

```python

# Make predictions on the test set
predictions = model.predict(x_test)

# Get the predicted class for each image
predicted_classes = [i.argmax() for i in predictions]


# Example: Display the first 10 images with their predicted labels
for i in range(10):
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Predicted: {predicted_classes[i]}, Actual: {y_test[i]}")
    plt.show()
```
